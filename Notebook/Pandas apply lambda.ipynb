```python
# Challenge 1:
def bin_function(x):
    if 0<x<1000: 
        i=1
    if 1000<=x<10000: 
        i=2
    if 10000<=x<100000: 
        i=3
    if 100000<=x<1000000: 
        i=4
    if x>=1000000:
        i=5
    return i
    
```

```python
df['bin']=df['Votes'].apply(bin_function)
```

```python
df.head(10)
```

```python
def rev_function(x,y):
    z=x*y
    return z
    
df['rev_min']=df.apply(lambda linea : rev_function( linea['Runtime (Minutes)'], linea['Revenue (Millions)']), axis=1 )


 
```

```python
def gen_function(gen,rk):
    if 'Comedy'in gen:
        return rk+1
    if 'Thriller'in gen:
        return rk-1
    else:
        return rk
```

```python
df['Rank2']=df.apply(lambda linea : gen_function(linea['Genre'],linea['Rank']), axis=1)
```
